These note is to define the samples of blazars and non-blazar objects for future data analysis.
Eventually, we will have a list of coordinates (RA,Dec), `ID` and *class* for each sample.

```python
def sex2deg(ra,dec):
    h,m,s = list(map(float,ra.split()))
    ra = s / 86400
    ra += m / 1440
    ra += h / 24
    ra *= 360
    
    d,m,s = dec.split()
    dec = float(s) / 3600
    dec += float(m) / 60
    if d[0] == '-':
        dec = float(d) - dec
    else:
        dec = float(d) + dec
    
    return ra,dec
```

```python
ra,dec = sex2deg('00 59 44.9','-00 21 30')

print(ra,dec)
```

```python
RA_COL = 'RA(2000)'
DEC_COL = 'DEC(2000)'

def apply_df(df):
    try:
        bla = list(sex2deg(df[RA_COL],df[DEC_COL]))
    except:
        bla = [None,None]
    return bla

agns[['RA','DEC']] = agns.apply(apply_df, axis=1, result_type='expand')
```

```python
agns.index.name = 'ID'
agns.reset_index(inplace=True)
agns
```

### Define the columns of interest

```python
col = dict(ra='RA',
           dec='DEC',
           idn='ID',
           name='Name',
           redshift='redsh',
           flux_opt='Vmag',
           flux_radio='Radio Fl',
           flux_xray='X Fl(mJ',
           klass='Class'
          )
```

```python
# Remove objects without radio flux
#
agns = agns.loc[agns[col['flux_radio']] > 0]


# Keep only the columns of interest
#
agns = agns[list(col.values())]

# And let's order our catalog by their Radio flux
#
agns = agns.sort_values(col['flux_radio'], ascending=False).reset_index(drop=True)

# Create a "label" column for running and labeling it in VOU-Blazas
#
agns['label'] = agns[col['idn']].astype(str) +'_'+ agns[col['ra']].round(4).astype(str).str.replace('.','_') +'_'+ agns[col['dec']].round(4).astype(str).str.replace('.','_')

columns = agns.columns
columns = [columns[i-1] for i in range(len(columns))]

agns = agns[columns]
```

## Recover results

To read the resulting SEDs, for each object, we may have to do some small processing on the RA/Dec values of each object to properly relate to the data/SED files.
The reason is the name convention used used in the output files:
* directories in `Results/` are named after the given/input RA and Dec values; on the other hand,
* directories in `Results/SEDtool/` are named after the object's (assumed) Radio source position.

To correlate those results to the original table, we may simply parse the (file/directory) in hands and truncate the decimal part to -- say -- 2 digits.

```python
ra_s1 = agns[col['ra']].astype(str)
ra_s1 = [s.split('.') for s in ra_s1]
ra_s1 = ['_'.join([l[0],l[1][:3]]) for l in ra_s1]

dec_s1 = agns[col['dec']].astype(str)
dec_s1 = [s.split('.') for s in dec_s1]
dec_s1 = ['_'.join([l[0],l[1][:3]]).replace('-','m') for l in dec_s1]

ids = agns['label']
```

```python
from os import path
from glob import glob
seds = [path.basename(p) for p in glob('work/AGN_all_radio_non_null/Results/SEDtool/*_sedtool.txt')]
seds_filename = seds[:]

seds = [s.replace('_sedtool.txt','') for s in seds]

seds = [s.split('_') for s in seds]

ra_s2 = ['_'.join([l[0],l[1][:3]]) for l in seds]

dec_s2 = ['_'.join([l[2],l[3][:3]]) for l in seds]
```

```python
df1 = pd.DataFrame(data={'ra_1':ra_s1, 'dec_1':dec_s1, 'ids':ids})

df2 = pd.DataFrame(data={'ra_2':ra_s2, 'dec_2':dec_s2, 'filenames':seds_filename})
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1['radec1'] = df1.apply(lambda x:'_'.join([x['ra_1'],x['dec_1']]), axis=1)
# df1.sort_values(['ra_1','dec_1'], inplace=True)
```

```python
df2['radec2'] = df2.apply(lambda x:'_'.join([x['ra_2'],x['dec_2']]), axis=1)
# df2.sort_values(['ra_2','dec_2'], inplace=True)
```

```python
df3 = df1.merge(df2, left_on='radec1', right_on='radec2', how='inner')
df3
```

```python
agns_results = agns.merge(df3[['ids','filenames']], left_on='label', right_on='ids')
agns_results
```
