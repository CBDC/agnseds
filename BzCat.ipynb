## 5BzCat: Blazars sample

```python
bzcat = pd.read_csv('5BZCAT.csv')
bzcat.sample(10)
```

### Define the columns of interest

```python
col = dict(ra = 'RA (J2000.0)',
           dec = 'Dec (J2000.0)',
           idn = 'id',
           name = 'Source name',
           flux_radio = 'Flux density 1.4/0.843GHz(mJy)',
           klass = 'Source classification',
           flux_opt = 'Rmag',
           flux_xray = 'X-ray flux 0.1-2.4 keV(1.e-12 cgs)',
           redshift = 'Redshift',
          )
```

```python
# Keep only the columns of interest

bzcat = bzcat[list(col.values())]
bzcat
```

```python
# Clean the Redshift column from '?'

bzcat[col['redshift']] = bzcat[col['redshift']].str.replace('\s*\?','')
```

```python
# And let's order our catalog by their Radio flux

bzcat = bzcat.sort_values(col['flux_radio'], ascending=False).reset_index(drop=True)
bzcat
```

```python
# Let's create two columns with the log of radio/xray fluxes

import numpy as np
bzcat['radio_log'] = np.log(bzcat[col['flux_radio']])
_ind = bzcat[col['flux_radio']] <= 0
bzcat.loc[_ind,'radio_log'] = None

bzcat['xray_log'] = np.log(bzcat[col['flux_xray']])
_ind = bzcat[col['flux_xray']] <= 0
bzcat.loc[_ind,'xray_log'] = None

sns.pairplot(bzcat[['radio_log','xray_log', col['flux_opt']]], diag_kind='kde', kind='scatter')
```

```python
del bzcat['xray_log'],bzcat['radio_log']
```

```python
bzcat['Label'] = bzcat[col['idn']].astype(str) +'_'+ bzcat[col['ra']].astype(str).str.replace('.','_') +'_'+ bzcat[col['dec']].astype(str).str.replace('.','_')
```

## Recover results

To read the resulting SEDs, for each object, we may have to do some small processing on the RA/Dec values of each object to properly relate to the data/SED files.
The reason is the name convention used used in the output files:
* directories in `Results/` are named after the given/input RA and Dec values; on the other hand,
* directories in `Results/SEDtool/` are named after the object's (assumed) Radio source position.

To correlate those results to the original table, we may simply parse the (file/directory) in hands and truncate the decimal part to -- say -- 2 digits.

```python
#ra_s1 = bzcat[col['ra']].apply(lambda x:'{:.5f}'.format(x)).str.replace('.','_')
ra_s1 = bzcat[col['ra']].astype(str)
ra_s1 = [s.split('.') for s in ra_s1]
ra_s1 = ['_'.join([l[0],l[1][:3]]) for l in ra_s1]

#dec_s1 = bzcat[col['dec']].apply(lambda x:'{:.5f}'.format(x)).str.replace('.','_').str.replace('-','m')
dec_s1 = bzcat[col['dec']].astype(str)
dec_s1 = [s.split('.') for s in dec_s1]
dec_s1 = ['_'.join([l[0],l[1][:3]]).replace('-','m') for l in dec_s1]

ids = bzcat['Label']
```

```python
from os import path
from glob import glob
seds = [path.basename(p) for p in glob('work/bzcat/Results/SEDtool/*_sedtool.txt')]
seds_filename = seds[:]

seds = [s.replace('_sedtool.txt','') for s in seds]

seds = [s.split('_') for s in seds]

ra_s2 = ['_'.join([l[0],l[1][:3]]) for l in seds]

dec_s2 = ['_'.join([l[2],l[3][:3]]) for l in seds]
```

```python
df1 = pd.DataFrame(data={'ra_1':ra_s1, 'dec_1':dec_s1, 'ids':ids})

df2 = pd.DataFrame(data={'ra_2':ra_s2, 'dec_2':dec_s2, 'filenames':seds_filename})
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1['radec1'] = df1.apply(lambda x:'_'.join([x['ra_1'],x['dec_1']]), axis=1)
# df1.sort_values(['ra_1','dec_1'], inplace=True)
```

```python
df2['radec2'] = df2.apply(lambda x:'_'.join([x['ra_2'],x['dec_2']]), axis=1)
# df2.sort_values(['ra_2','dec_2'], inplace=True)
```

```python
df3 = df1.merge(df2, left_on='radec1', right_on='radec2', how='inner')
df3
```

```python
bzcat_results = bzcat.merge(df3[['ids','filenames']], left_on='Label', right_on='ids')
bzcat_results
```
